```python
#reduce the level of display type
a=[]
for i in df['displayType']:
    if i==' Full HD+ Display' or i ==' HD+ Display' or i==' Full HD Display' or i==' HD Display': 
        a.append('HD_DISPLAY')
    elif i==' Full HD+ Super AMOLED Display' or i==' Full HD+ E3 Super AMOLED Display':
            a.append('SUPER_AMOLED')
    elif i==' Display':
                a.append('Display')
    elif i==' Full HD+ AMOLED Display':
                    a.append('HD_AMOLED')
    elif i==' Quad HD+ Display' or i==' Quad HD Display':
                        a.append('QUAD_HD')
    elif i==' QXGA+ Display':
                            a.append('QXGA')
    else:
        a.append(i)
df['displayType']=a
```

```python
#collect data of memory capacity
new = df["ram"].str.split("|", n = 2, expand = True)
df['RAM']=new[0]
df['ROM']=new[1]
df['expandabe']=new[2]

#extract numeric from alpha numeric and convert into int datatype
df['RAM']=df['RAM'].str.extract(r'^(-?\d+)',expand=False).astype(int)
df["ROM"]=df["ROM"].replace(" GB ROM","",regex=True).replace(" MB ROM ","",regex=True).replace(" ","",regex=True).astype(int)
```

```python
#convert TB to GB
df["expandabe"]=df["expandabe"].replace("Expandable Upto ","",regex=True)
df["expandabe"]=df["expandabe"].replace("1 TB","1000 GB",regex=True)
df["expandabe"]=df["expandabe"].replace("2 TB","2000 GB",regex=True)
df["expandabe"] = df["expandabe"].str.replace('\D+', '')
df=df[df['expandabe']!='7']
```

```python
#taking first word of processor
new = df["Processor"].str.split(" ", n = 1, expand = True)
df["process"]=new[0]
df.process=df.process.apply(str.upper)

#reduce the levels in preocess column
a=[]
for i in df['process']:
    if 'QUALCOMM' in i or "SNAPDRAGON" in i: 
        a.append('QUALCOMM')
    elif 'HELIO' in i or "MTK" in i or 'DIMENSITY' in i:
        a.append('MEDIATEK')
    elif 'OCTA' in i:
        a.append('OCTA')
    elif 'OTHERS' in i:
        a.append('OTHERS')
    else:
        a.append('others1')
df.process=a
df=df.drop(columns=['Processor'])
```

```python
#extract camera details 
new = df["Camera"].str.split("|", n = 1, expand = True)

#camera caolumn have the information of front and rare camera
df['rareCamera']=new[0]
df['FrontCamera']=new[1]
```

```python
#mention only max MP in frontcamera
new = df["FrontCamera"].str.split("MP", n = 1, expand = True)
new.iloc[:,1].value_counts()
df['frontMP']=new[0].astype(float)
#if there is no camera in front, it asign to zero
df["frontMP"]=df["frontMP"].fillna(0)

#In frontDual column, we only mentioning-phone having front dual camera or not
df['frontDual']=new[1]
df["frontDual"] = df["frontDual"].str.replace('\d+', '').str.replace(" ","")


#if phone having front dual caemra it asign one, otherwise zero 
frontDual=[]
for i in df['frontDual']:
    if i=='+MPDualFrontCamera' or i=='DualFrontCamera' :
        frontDual.append(1)
    else:
        frontDual.append(0)
df["frontDual"]=frontDual
df["frontDual"]=df["frontDual"].astype(int)
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,4,7])
],remainder='passthrough')


step2 =  LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train.values,y_train)
y_pred = pipe.predict(X_test.values)


print('R2 score for train',pipe.score(X_train,y_train))
print('R2 score for test',pipe.score(X_test,y_test))
print('MAE',mean_absolute_error(y_test,y_pred))

model.append('LinearRegression')
r2_score_train.append(pipe.score(X_train,y_train))
r2_score_test.append(pipe.score(X_test,y_test))
```
